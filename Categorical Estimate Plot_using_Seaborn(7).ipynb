### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

```python

import seaborn as sns
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

# Bar_plot

```python
import numpy as np
```

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# barplot
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.std)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.max)
```

# Point_plot

```python
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',ci=None)
```

# countplot

```python
sns.countplot(data=tips,x='sex')
```

```python

sns.countplot(data=tips,x='sex',hue='day')
```

- faceting using catplot

```python
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```
