#  Facet plots 

```python

import seaborn as sns
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
colors = ["#00CD00", "#00FFAA", "#FCCC00", "#FF0000"]
colors = sns.color_palette(colors)
sns.palplot(colors)
plt.show()
```

```python
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="day", height=3.5, aspect=.65)
g.map(sns.histplot, "total_bill")
```

# Pairgrid and pairplot

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')

g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.catplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
