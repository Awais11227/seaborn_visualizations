## Categorical Plots

### Categorical Scatter Plot

- Stripplot
- Swarmplot



```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(data=tips, x='total_bill',y='tip')
```

```python
# strip plot -> axes level function

sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
# using catplot

sns.catplot(data=tips, x='day',y='total_bill',kind='strip',hue='sex')
```

```python
# swarmplot
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm',color='red')
```

```python
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex')
```
